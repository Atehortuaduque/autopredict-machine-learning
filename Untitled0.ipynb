# GN Bayes

```python
new_d=data.drop(['name','fuel','selling_price','transmission',"owner"],axis=1)

new_d.dropna(inplace=True)

new_d.head(10)
```

```python
X = new_d.values[:,:-1]
y = new_d.values[:,-1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=21)
from sklearn.naive_bayes import GaussianNB
est = GaussianNB()
est.fit(X_train,y_train)
print(accuracy_score(est.predict(X_test), y_test))

GN_bayes_accuracy = accuracy_score(est.predict(X_test), y_test)
```

# Decision Tree


```python
new_d=data.drop(['name','fuel','selling_price','transmission',"owner"],axis=1)

new_d.dropna(inplace=True)

new_d.head(10)
```

```python
#decision Tree
X = new_d.values[:,:-1]
y = new_d.values[:,-1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=21)
from sklearn.tree import DecisionTreeClassifier
est = DecisionTreeClassifier(max_depth=2)
est.fit(X_train,y_train)
print(accuracy_score(est.predict(X_test), y_test))

decision_tree_accuracy = accuracy_score(est.predict(X_test), y_test)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X = new_d.values[:, :-1]
y = new_d.values[:, -1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=21)

max_depth_values = range(1, 11)
accuracy_values = []

for max_depth in max_depth_values:
    est = DecisionTreeClassifier(max_depth=max_depth)
    est.fit(X_train, y_train)

    predictions = est.predict(X_test)

    accuracy = accuracy_score(y_test, predictions)
    accuracy_values.append(accuracy)

plt.plot(max_depth_values, accuracy_values, marker='o')
plt.xlabel('max_depth')
plt.ylabel('Accuracy')
plt.title('Accuracy vs. max_depth')
plt.show()
```

# Random Forest


```python
new_d=data.drop(['name','fuel','selling_price','transmission',"owner"],axis=1)
new_d.dropna(inplace=True)
new_d.head(10)
```

```python
X = new_d.values[:,:-1]
y = new_d.values[:,-1]
from sklearn.ensemble import RandomForestClassifier
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=17)

est = RandomForestClassifier()
est.fit(X_train,y_train)
print(accuracy_score(est.predict(X_test), y_test))
accuracy_values=[]
for max_depth in range(1,15):
    est = DecisionTreeClassifier(max_depth=max_depth,criterion="entropy")
    est.fit(X_train, y_train)

    predictions = est.predict(X_test)

    accuracy = accuracy_score(y_test, predictions)
    accuracy_values.append(accuracy)
plt.style.use('seaborn-darkgrid')
plt.plot(range(1,15), accuracy_values, marker='o')
plt.xlabel('max_depth')
plt.ylabel('Accuracy')
plt.title('Accuracy vs. max_depth')
plt.show()

Random_forest_accuracy = accuracy_score(est.predict(X_test), y_test)
```

```python
accuracy_values = []
for n_estimators in range(1, 10):
    est = RandomForestClassifier(n_estimators=n_estimators, max_depth=8, random_state=17)
    est.fit(X_train, y_train)

    predictions = est.predict(X_test)

    accuracy = accuracy_score(y_test, predictions)
    accuracy_values.append(accuracy)
plt.style.use('seaborn-darkgrid')
plt.plot(range(1, 10), accuracy_values, marker='o')
plt.xlabel('n_estimators')
plt.ylabel('Accuracy')
plt.title('Accuracy vs. n_estimators')
plt.show()
```

```python
accuracy_values = []
for x in range(1, 10):
    est = RandomForestClassifier(n_estimators=x, max_depth=x, random_state=17)
    est.fit(X_train, y_train)

    predictions = est.predict(X_test)

    accuracy = accuracy_score(y_test, predictions)
    accuracy_values.append(accuracy)
plt.style.use('seaborn-darkgrid')
plt.plot(range(1, 10), accuracy_values, marker='o')
plt.xlabel('n_estimators')
plt.ylabel('Accuracy')
plt.title('Accuracy vs. n_estimators')
plt.show()
```
